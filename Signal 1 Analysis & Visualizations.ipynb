## Now let's see what the 'COP' movement path looked like for the individual...
### We will take a '30-point average' (i.e., average every 30 rows) so that we can shrink the data from 30,000 to 1,000...which will allow for us to better see variations over time within a line graph

```python
cop_point_avg = 30
cop_list = []

def point_avg(data):
    data_sum = 0
    count = 0
    for num in data:
        data_sum += num
        count += 1
        if count == cop_point_avg:
            cop_list.append(data_sum / cop_point_avg)
            data_sum = 0
            count = 0
    return cop_list

cop_x_30 = point_avg(df['COP:X'])
print(cop_x_30)
```

```python
plt.figure(figsize = (12,12))
plt.plot(cop_x_30, range(len(cop_x_30)), linewidth = 3.5)
plt.grid(axis = 'x', alpha = 0.3)
plt.title('Center of Pressure M/L Displacement (1000Hz)', fontsize = 20, fontweight = 'bold')
plt.yticks([0.0, (5/30) * float(len(cop_x_30)),
              (10/30) * float(len(cop_x_30)), (15/30) * float(len(cop_x_30)),
              (20/30) * float(len(cop_x_30)), (25/30) * float(len(cop_x_30)),
              (1.0) * float(len(cop_x_30))], ['0sec.', '5sec.', '10sec.', '15sec.', '20sec.', '25sec.', 
                                              '30sec.'], fontweight = 'bold')
plt.ylabel('Time', fontsize = 12, fontweight = 'bold')
plt.xticks([200, 210, 220, 230, 240, 250, 260, 270, 280],
           ['200mm', '210mm', '220mm', '230mm', '240mm', '250mm', '260mm', '270mm', '280mm'], fontweight = 'bold')
plt.xlabel('Medial/Lateral Movement', fontsize = 12, fontweight = 'bold')
sns.despine()
plt.show()
```

```python
cop_y_30 = point_avg(df['COP:Y'])
print(cop_y_30)
```

```python
plt.figure(figsize = (15,10))
plt.plot(range(len(cop_y_30)), cop_y_30, linewidth = 3.5)
plt.grid(axis = 'y', alpha = 0.3)
plt.title('Center of Pressure A/P Displacement (1000Hz)', fontsize = 20, fontweight = 'bold')
plt.xlabel('Time', fontsize = 12, fontweight = 'bold')
plt.xticks([0.0, (5/30) * float(len(cop_y_30)),
              (10/30) * float(len(cop_y_30)), (15/30) * float(len(cop_y_30)),
              (20/30) * float(len(cop_y_30)), (25/30) * float(len(cop_y_30)),
              (1.0) * float(len(cop_y_30))], ['0sec.', '5sec.', '10sec.', '15sec.', '20sec.', '25sec.', '30sec.'], 
          fontweight = 'bold')
plt.ylabel('Anterior/Posterior Movement', fontsize = 12, fontweight = 'bold')
plt.yticks([200, 220, 240, 260, 280, 300, 320, 340], ['200mm', '220mm', '240mm', '260mm', '280mm', '300mm', '320mm', '340mm'],
          fontweight = 'bold')
sns.despine()
plt.show()
```

```python
# Force:Z Analysis (Force produced in the Z Direction...also known as the vertical direction)

force_z_mean = df['Force:Z'].mean()
force_z_mean = round(force_z_mean, 2)
print('Force:Z Mean = ', force_z_mean, 'N')

force_z_median = df['Force:Z'].median()
force_z_median = round(force_z_median, 2)
print('Force:Z Median = ', force_z_median, 'N')

force_z_min = df['Force:Z'].min()
force_z_min = round(force_z_min, 2)
print('Force:Z Min = ', force_z_min, 'N')

force_z_max = df['Force:Z'].max()
force_z_max = round(force_z_max, 2)
print('Force:Z Max = ', force_z_max, 'N')

force_z_range = (df['Force:Z'].max()) - (df['Force:Z'].min())
force_z_range = round(force_z_range, 2)
print('Force:Z Range = ', force_z_range, 'N')

force_z_std = df['Force:Z'].std()
force_z_std = round(force_z_std, 2)
print('Force:Z Std = ', force_z_std, 'N')
```

### When it comes to measuring the Z Force during a quiet stance, it is usually only useful when performing more complex calculations or more complex movements than a 30sec. quiet stance. But we can use it to determine the weight of the individual that this data was taken from.

```python
weight_kg = force_z_mean / 9.81  
# 9.81 = Force from gravity
weight_lbs = weight_kg * 2.2
print("Subject's bodyweight = ", round(weight_kg, 2), 'kg', ' or ', round(weight_lbs, 2), 'lbs.')
```

### Lastly, we will take a look to see how many outliers are present. We can use the center of pressure data since they have the largest standard deviations and ranges (max/min differences). This may be due to the use 'mm' instead of 'cm' for the units of measure, but we will check anyways to be sure. This is simply just a demonstration of one way to do it manually.

```python
x_upper_limit = cop_x_mean + (3 * cop_x_std)
x_lower_limit = cop_x_mean - (3 * cop_x_std)

print('COP:X Upper Limit = ', round(x_upper_limit, 2), 'mm')
print('COP:X Lower Limit = ', round(x_lower_limit, 2), 'mm')
```

```python
for i in df['COP:X']:
    if i > x_upper_limit:
        print(i)
# This for loop did not return any values, because 'upper_limit' is larger than
# cop_x_max, which = 279.88.
```

```python
num_lower_outliers = []
for i in df['COP:X']:
    if i < x_lower_limit:
        num_lower_outliers.append(i)
print('# of Outliers Observed Past the Lower Limit = ', len(num_lower_outliers))

# According to this, there are 277 out of 30,000+ rows with values considered to
# be outliers
```

```python
y_upper_limit = cop_y_mean + (3 * cop_y_std)
y_lower_limit = cop_y_mean - (3 * cop_y_std)

print('COP:Y Upper Limit = ', round(y_upper_limit, 2), 'mm')
print('COP:Y Lower Limit = ', round(y_lower_limit, 2), 'mm')

# Becuase we know that both the max (344.69mm) & min (291.49mm) values for
# df[COP:Y] are both within the upper & lower limits, we already know that
# there are no outliers in this column.
```
