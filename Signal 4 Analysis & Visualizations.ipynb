# Double-Leg Quiet Stance
## Analysis of raw Force Plate Data - 60sec. @ 1,000Hz

```python
# Import libraries that may be needed
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Force Plate - Signal 4.csv")
print(df.head())
```

```python
df.info()
```

### 1.) We can assume that this data was collected at 1,000Hz (60,000 rows / 60sec. = 1,000Hz)
### 2.) We will only be analyzing columns 0 & 1 (COP X & Y) and 6-8 (Force X, Y, & Z)

## <u>Comparison Between Center of Pressure (X,Y) Data:</u>

```python
# COP:X Analysis (Center of Pressure in the X Direction...also known as medio-lateral)
# mm = millimeters

cop_x_mean = df['COP:X'].mean()
print('COP:X Mean = ', round(cop_x_mean, 2), 'mm')

cop_x_median = df['COP:X'].median()
print('COP:X Median = ', round(cop_x_median, 2), 'mm')

cop_x_min = df['COP:X'].min()
print('COP:X Min = ', round(cop_x_min, 2), 'mm')

cop_x_max = df['COP:X'].max()
print('COP:X Max = ', round(cop_x_max, 2), 'mm')

cop_x_range = (cop_x_max) - (cop_x_min)
print('COP:X Range = ', round(cop_x_range, 2), 'mm')

cop_x_std = df['COP:X'].std()
print('COP:X Std = ', round(cop_x_std, 2), 'mm')
```

```python
plt.figure(figsize = (18, 10))
sns.violinplot(data = df['COP:X'], orient = 'h')
plt.grid(axis = 'x', alpha = 0.3)
plt.xlabel('Total Movement Observed', fontsize = 12, fontweight = 'bold')
plt.xticks([190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290],
           ['190mm', '200mm', '210mm', '220mm', '230mm', '240mm', '250mm', '260mm', '270mm', '280mm', '290mm'], 
           fontweight = 'bold')
plt.yticks([0], [])
plt.title('Medial/Lateral COP Data Distribution (60sec.)', fontsize = 20, fontweight = 'bold')
sns.despine()
plt.show()
```

```python
# COP:Y Analysis (Center of Pressure in the Y Direction...also known as anterio-posterior)

cop_y_mean = df['COP:Y'].mean()
print('COP:Y Mean = ', round(cop_y_mean, 2), 'mm')

cop_y_median = df['COP:Y'].median()
print('COP:Y Median = ', round(cop_y_median, 2), 'mm')

cop_y_min = df['COP:Y'].min()
print('COP:Y Min = ', round(cop_y_min, 2), 'mm')

cop_y_max = df['COP:Y'].max()
print('COP:Y Max = ', round(cop_y_max, 2), 'mm')

cop_y_range = (cop_y_max) - (cop_y_min)
print('COP:Y Range = ', round(cop_y_range, 2), 'mm')

cop_y_std = df['COP:Y'].std()
print('COP:Y Std = ', round(cop_y_std, 2), 'mm')
```

```python
plt.figure(figsize = (10, 15))
sns.violinplot(data = df['COP:Y'], orient = 'v')
plt.grid(axis = 'y', alpha = 0.3)
plt.title('Anterior/Posterior COP Data Distribution (60sec.)', fontsize = 20, fontweight ='bold')
plt.yticks([150, 200, 250, 300, 350, 400, 450, 500], ['150mm', '200mm', '250mm', '300mm', '350mm', '400mm', '450mm', '500mm'], 
           fontweight = 'bold')
plt.ylabel('Total Movement Observed', fontsize = 12, fontweight = 'bold')
plt.xticks([0], [])
sns.despine()
plt.show()
```

## Now let's see what the 'COP' movement path looked like for the individual...
### We will take a '60-point average' (i.e., average every 30 rows) so that we can shrink the data from 60,000 to 1,000...which will allow for us to better see variations over time within a line graph

```python
cop_point_avg = 60
cop_list = []

def point_avg(data):
    data_sum = 0
    count = 0
    for num in data:
        data_sum += num
        count += 1
        if count == cop_point_avg:
            cop_list.append(data_sum / cop_point_avg)
            data_sum = 0
            count = 0
    return cop_list

cop_x_60 = point_avg(df['COP:X'])
```

```python
plt.figure(figsize = (10,10))
plt.plot(cop_x_60, range(len(cop_x_60)), linewidth = 3.5)
plt.grid(axis = 'x', alpha = 0.3)
plt.title('Center of Pressure M/L Displacement (1000Hz)', fontsize = 20, fontweight = 'bold')
plt.yticks([0.0, (5/60) * float(len(cop_x_60)), (10/60) * float(len(cop_x_60)), (15/60) * float(len(cop_x_60)),
            (20/60) * float(len(cop_x_60)), (25/60) * float(len(cop_x_60)), (30/60) * float(len(cop_x_60)), 
            (35/60) * float(len(cop_x_60)), (40/60) * float(len(cop_x_60)), (45/60) * float(len(cop_x_60)),
              (50/60) * float(len(cop_x_60)), (55/60) * float(len(cop_x_60)), (1.0) * float(len(cop_x_60))], 
           ['0sec.', '5sec.', '10sec.', '15sec.', '20sec.', '25sec.', '30sec.', '35sec.', '40sec.', '45sec.', 
            '50sec.', '55sec.', '60sec.'], fontweight = 'bold')
plt.ylabel('Time', fontsize = 12, fontweight = 'bold')
plt.xticks([210, 220, 230, 240, 250, 260, 270, 280, 290],
           ['210mm', '220mm', '230mm', '240mm', '250mm', '260mm', '270mm', '280mm', '290mm'], fontweight = 'bold')
plt.xlabel('Medial/Lateral Movement', fontsize = 12, fontweight = 'bold')
sns.despine()
plt.show()
```

```python
cop_y_60 = point_avg(df['COP:Y'])

plt.figure(figsize = (15,10))
plt.plot(range(len(cop_y_60)), cop_y_60, linewidth = 3.5)
plt.grid(axis = 'y', alpha = 0.3)
plt.title('Center of Pressure A/P Displacement (1000Hz)', fontsize = 20, fontweight = 'bold')
plt.xlabel('Time', fontsize = 12, fontweight = 'bold')
plt.xticks([0.0, (5/60) * float(len(cop_y_60)), (10/60) * float(len(cop_y_60)), (15/60) * float(len(cop_y_60)),
            (20/60) * float(len(cop_y_60)), (25/60) * float(len(cop_y_60)), (30/60) * float(len(cop_y_60)), 
            (35/60) * float(len(cop_y_60)), (40/60) * float(len(cop_y_60)), (45/60) * float(len(cop_y_60)),
              (50/60) * float(len(cop_y_60)), (55/60) * float(len(cop_y_60)), (1.0) * float(len(cop_y_60))], 
           ['0sec.', '5sec.', '10sec.', '15sec.', '20sec.', '25sec.', '30sec.', '35sec.', '40sec.', '45sec.', 
            '50sec.', '55sec.', '60sec.'], fontweight = 'bold')
plt.ylabel('Anterior/Posterior Movement', fontsize = 12, fontweight = 'bold')
plt.yticks([150, 200, 250, 300, 350, 400, 450, 500], ['150mm', '200mm', '250mm', '300mm', '350mm', '400mm',
            '450mm', '500mm'], fontweight = 'bold')
sns.despine()
plt.show()
```

## <u>Comparison Between of Force (X, Y, Z) Data:</u>

```python
# Force:X Analysis (Force produced in the X Direction)
# N = Newtons

force_x_mean = df['Force:X'].mean()
print('Force:X Mean = ', round(force_x_mean, 2), 'N')

force_x_median = df['Force:X'].median()
print('Force:X Median = ', round(force_x_median, 2), 'N')

force_x_min = df['Force:X'].min()
print('Force:X Min = ', round(force_x_min, 2), 'N')

force_x_max = df['Force:X'].max()
print('Force:X Max = ', round(force_x_max, 2), 'N')

force_x_range = (force_x_max) - (force_x_min)
print('Force:X Range = ', round(force_x_range, 2), 'N')

force_x_std = df['Force:X'].std()
print('Force:X Std = ', round(force_x_std, 2), 'N')
```

```python
plt.figure(figsize = (18, 10))
sns.violinplot(data = df['Force:X'], orient = 'h')
plt.grid(axis = 'x', alpha = 0.3)
plt.xlabel('Total Force Distribution Observed', fontsize = 12, fontweight = 'bold')
plt.xticks([-12.5, -10, -7.5, -5, -2.5, 0, 2.5, 5, 7.5, 10, 12.5],
           ['-12.5N', '-10N', '-7.5N', '-5N', '-2.5N', '0N', '2.5N', '5N', '7.5N', '10N', '12.5N'], fontweight = 'bold')
plt.yticks([0], [])
plt.title('Medial/Lateral Force Data Distribution (60sec.)', fontsize = 20, fontweight = 'bold')
sns.despine()
plt.show()
```

```python
# Force:Y Analysis (Force produced in the Y Direction)

force_y_mean = df['Force:Y'].mean()
force_y_mean = round(force_y_mean, 2)
print('Force:Y Mean = ', force_y_mean, 'N')

force_y_median = df['Force:Y'].median()
force_y_median = round(force_y_median, 2)
print('Force:Y Median = ', force_y_median, 'N')

force_y_min = df['Force:Y'].min()
force_y_min = round(force_y_min, 2)
print('Force:Y Min = ', force_y_min, 'N')

force_y_max = df['Force:Y'].max()
force_y_max = round(force_y_max, 2)
print('Force:Y Max = ', force_y_max, 'N')

force_y_range = (force_y_max) - (force_y_min)
force_y_range = round(force_y_range, 2)
print('Force:Y Range = ', force_y_range, 'N')

force_y_std = df['Force:Y'].std()
force_y_std = round(force_y_std, 2)
print('Force:Y Std = ', force_y_std, 'N')
```

```python
plt.figure(figsize = (10, 16))
sns.violinplot(data = df['Force:Y'], orient = 'v')
plt.grid(axis = 'y', alpha = 0.3)
plt.ylabel('Total Force Distribution Observed', fontsize = 12, fontweight = 'bold')
plt.yticks([-60, -40, -20, 0, 20, 40, 60],
           ['-60N', '-40N', '-20N', '0N', '20N', '40N', '60N'], fontweight = 'bold')
plt.xticks([0], [])
plt.title('Anterior/Posterior Force Data Distribution (30sec.)', fontsize = 20, fontweight = 'bold')
sns.despine()
plt.show()
```

```python
# Force:Z Analysis (Force produced in the Z Direction...also known as the vertical direction)

force_z_mean = df['Force:Z'].mean()
force_z_mean = round(force_z_mean, 2)
print('Force:Z Mean = ', force_z_mean, 'N')

force_z_median = df['Force:Z'].median()
force_z_median = round(force_z_median, 2)
print('Force:Z Median = ', force_z_median, 'N')

force_z_min = df['Force:Z'].min()
force_z_min = round(force_z_min, 2)
print('Force:Z Min = ', force_z_min, 'N')

force_z_max = df['Force:Z'].max()
force_z_max = round(force_z_max, 2)
print('Force:Z Max = ', force_z_max, 'N')

force_z_range = (df['Force:Z'].max()) - (df['Force:Z'].min())
force_z_range = round(force_z_range, 2)
print('Force:Z Range = ', force_z_range, 'N')

force_z_std = df['Force:Z'].std()
force_z_std = round(force_z_std, 2)
print('Force:Z Std = ', force_z_std, 'N')
```

### When it comes to measuring the Z Force during a quiet stance, it is usually only useful when performing more complex calculations or more complex movements than a 60sec. quiet stance. But we can use it to determine the weight of the individual that this data was taken from.

```python
weight_kg = force_z_mean / 9.81  
# 9.81 = Force from gravity
weight_lbs = weight_kg * 2.2
print("Subject's bodyweight = ", round(weight_kg, 2), 'kg', ' or ', round(weight_lbs, 2), 'lbs.')
```
